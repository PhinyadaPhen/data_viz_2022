##Data 1 alldb.csv

```python
data1_alldb = pd.read_csv(os.path.join(path,'alldb.csv'))
data1_alldb ## แสดงตาราง
```

```python
data1_alldb= data1_alldb.rename(columns={'title':'gender'})  #  เปลี่ยนชื่อ คอลัมน์ title เป็น gender
data1_alldb ## แสดงตาราง
```

# ▶ user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)




```python
data1_alldb = data1_alldb[(data1_alldb['user'].astype(str).str.isdigit()) & (data1_alldb['user'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ user ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
data1_alldb ## แสดงตาราง
```

# ▶ RowKey เป็น String

```python
data1_alldb['RowKey']=data1_alldb['RowKey'].astype('string') #เปลี่ยน dtype เป็น  String
data1_alldb ## แสดงตาราง
```

```python
print(data1_alldb.dtypes)  #พบว่า RowKey เป็น String
```

# ▶ date เป็น YYYY-MM-DD

```python
data1_alldb['date'] = pd.to_datetime(data1_alldb ['date'],format='%Y/%m/%d') #เปลี่ยนรูปแบบของ วันเดือนปี 
data1_alldb ## แสดงตาราง
```

```python
data1_alldb[['date']] = data1_alldb[['date']].fillna('ไม่ระบุ') #แทน 'ไม่ระบุ' ในค่าว่าง
data1_alldb ## แสดงตาราง
```

# ▶ site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data1_alldb= data1_alldb[(data1_alldb['site'].astype(str).str.isdigit()) & (data1_alldb['site'].str.len() == 5)]         # เลือกมาเฉพาะ คอลัมน์ site ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
data1_alldb ## แสดงตาราง
```

# ▶ time เป็น hh-mm-ss

```python
data1_alldb['time'] = pd.to_datetime(data1_alldb['time'],format='%H:%M%:%S',errors='ignore')  #เปลี่ยนรูปแบบของ time 
data1_alldb ## แสดงตาราง
```

```python
data1_alldb[['time']] = data1_alldb[['time']].fillna('ไม่ระบุ')  #แทน ไม่ระบุ ในค่าว่าง
data1_alldb ## แสดงตาราง
```

# ▶ urine มีค่า String

```python
data1_alldb['urine']=data1_alldb['urine'].astype('string') #เปลี่ยน dtype เป็น  String
data1_alldb ## แสดงตาราง
```

```python
print(data1_alldb.dtypes) #พบว่า urine  เป็น   string
```

```python
data1_alldb[['urine']] = data1_alldb[['urine']].fillna('ไม่ระบุ')  #แทน ไม่ระบุ ในค่าว่าง
data1_alldb ## แสดงตาราง
```

# ▶ ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
data1_alldb= data1_alldb[(data1_alldb['PartitionKey'].astype(str).str.isdigit()) & (data1_alldb['PartitionKey'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ PartitionKey ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
data1_alldb ## แสดงตาราง
```

##Data 2  hospital_list.xlsx 

```python
data2_hospital = pd.read_excel(os.path.join(path,'hospital_list.xlsx')) #การนำเข้าไฟล์ที่มีตัวแปรคือdata2_hospital
data2_hospital ## แสดงตาราง
```

```python
data2_hospital = data2_hospital[['hcode','zone_code','name','province','amphur','tambon']] ##เลือกเฉพาะcolumnที่ต้องการ
data2_hospital  ## แสดงตาราง
```
