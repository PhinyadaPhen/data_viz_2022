# โหลดข้อมูลการตรวจโรค ov

```python
data_file_path = os.path.join(path,'alldb.csv') # เอาที่อยู่ของไฟล์เชื่อมกับที่อยู่ของไฟล์เรา โดยกำหนดชื่่อไฟล์ที่เราต้องการ
peopleov = pd.read_csv(data_file_path)
peopleov 
```

```python
peopleov1 = peopleov[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey','age']]
peopleov1
```

```python
peopleov1.isnull().any()
```

```python
peopleov1 = peopleov1.dropna()
peopleov1
```

```python
peopleov1.isnull().any()
```

```python
removes = peopleov.shape[0] - peopleov1.dropna().shape[0]
print(f'size before drop = {peopleov.shape[0]}')
print(f'size after drop = {peopleov1.dropna().shape[0]}')
print(f'we lost {100*(removes/peopleov1.shape[0])}% of data')
```

## แทนค่า missing ของ user ด้วยค่า class ใหม่(ไม่ระบุ)

```python
peopleov = peopleov.drop()
peopleov
```

```python
peopleov[['user']] = peopleov[['user']].fillna('')
```
